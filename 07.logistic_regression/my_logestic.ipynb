```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plt.rcParams.update({'font.size':16})
t= np.linspace(-5,5,100)
sig = 1/(1+np.exp(-t)) 
plt.plot(t,sig,'b-')
plt.xlabel('t')
plt.ylabel('$\sigma(t)$')
plt.grid()
```

# Built-in function in python

```python
from scipy.special import expit
# res=expit([-np.inf, 0,np.inf])
res = expit(np.linspace(-5,5,100))
```

```python
plt.plot(res)
```

# Derivative:

```python
t = np.linspace(-5,5,100)
sig = expit(t)
sig_prime = expit(t) *(1- expit(t))
plt.plot(t,sig,'b-',label ="$\sigma(t)$")
plt.plot(t,sig_prime,'r--',label='$\sigma^{\prime}(t)$')
plt.xlabel('t')
plt.legend()
plt.grid()
```
